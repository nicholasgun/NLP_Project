```python
import pandas as pd
import numpy as np
```

Untuk data training model bert sebagai classifier sentiment dari sebuah aspek, data review tokopedia yang sudah dilabel akan diubah menjadi berikut:
- Review (text)
- Aspek (text {'pelayanan', 'pengiriman', 'barang'})
- Sentimen (0: negatif, 1: netral, 2:positif)

```python
df = pd.read_csv('tokopedia_review - tokopedia_review.csv')
df.head()
```

```python
df.shape
```

```python
df = df.drop(list(df.columns[5:]), axis=1)
df.head()
```

```python
unpivot = pd.melt(df, id_vars=['review', 'rating'], value_vars=['pelayanan', 'pengiriman', 'barang'])
unpivot
```

```python
dropped = unpivot.dropna(axis=0, how='any')
print(dropped.shape)
dropped
```

```python
dropped.to_csv('out.csv', index=False)
```
